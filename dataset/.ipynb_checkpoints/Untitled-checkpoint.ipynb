```python
import os    
import csv

all_files = os.listdir("labels/")

def count_lines(f):
    line_count = 0
    for line in f:
        if line != "\n":
            line_count += 1
    return line_count

s = 0
t = 0

with open("img_less_people.csv", "w") as f:
    writer = csv.writer(f)
    for file in all_files:
        if file.endswith(".txt"):
            file_path = f"labels/{file}"
            txt = open(file_path, "r")
            line_count = 0
            for line in txt:
                if line != "\n":
                    line_count += 1
            if line_count > 5:
                s += 1
            else:
                print(file)
                writer.writerow([file])
            t += 1
            txt.close()


print(s,t)
```
